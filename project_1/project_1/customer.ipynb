```python
df2 = pd.get_dummies(df).astype(int)
df2.head()
```

```python
df2 = pd.get_dummies(df,drop_first='if_binary').astype(int)
df2.head()
```

```python
x = df2.drop('satisfaction_satisfied',axis=1)
y=df2[['satisfaction_satisfied']]
```

```python
#train and split
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
#decision tree
```

```python
from sklearn.tree import DecisionTreeClassifier #classification
from sklearn.tree import DecisionTreeRegressor  #regresion
from sklearn.tree import plot_tree
```

```python
dtc = DecisionTreeClassifier()
dtc.fit(x_train,y_train)
```

```python
# plot_tree(dtc)

print(dtc.score(x_train,y_train))
print(dtc.score(x_test,y_test))
```

```python
#DecisionTreeRegressor
#prediction 
# confusion matrix 
# classification matrix 
# model save 
```

```python
#model training 
# algorithm training
#classification  --> logicticregression 

from sklearn.linear_model import LogisticRegression 

```

```python
lg = LogisticRegression()
```

```python
lg.fit(x_train,y_train)
```

```python
print(lg.score(x_train,y_train))
print(lg.score(x_test,y_test))
```

```python
#model test 
#confusion matrix 


from sklearn.metrics import confusion_matrix , classification_report
```

```python
#prediction 
pred = lg.predict(x_test)
```

```python
pred
```

```python
y_test['prediction'] = pred
```
